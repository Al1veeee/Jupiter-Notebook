```python
#1.1.1 –ù–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã –≤ –ø–æ–ø—É–ª—è—Ü–∏–∏ –∏ –æ–±—â–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ–ø—É–ª—è—Ü–∏–∏ –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä–µ—Ö —ç—Ç–∞–ø–æ–≤ —ç–≤–æ–ª—é—Ü–∏–∏.

def qZ(x, y): # –§—É–Ω–∫—Ü–∏—è –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º
    return (x - 3 * y + 1) / (3 * x ** 2 + 3 * y ** 2 + 1)

def qSumZ(Z): # –°—É–º–º–∞ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º
    return sum(Z)
#—Å—Ö–µ–º–∞ —Ö—Ä–æ–º–æ—Å–æ–º
def exchangeSheme (oldX, oldY, sortedId):
    X = [0 for i in range(4)]
    Y = [0 for i in range(4)]

    X[2] = oldX[sortedId[2]]
    X[3] = oldX[sortedId[2]]

    X[0] = oldX[sortedId[0]]

    X[1] = oldX[sortedId[1]]

    Y[0] = oldY[sortedId[2]]
    Y[1] = oldY[sortedId[2]]

    Y[2] = oldY[sortedId[0]]

    Y[3] = oldY[sortedId[1]]

    return X, Y

def sorting(Z): # –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –ø–æ—Ç–æ–º–∫–æ–≤ –∏ –≤—ã–¥–µ–ª–∏—Ç—å –∏–Ω–¥–µ–∫—Å—ã
    sortedId = sorted(range(len(Z)), key = lambda k: Z[k])

    return sortedId

def evoStep(X, Y, Z): # –®–∞–≥ —ç–≤–æ–ª—é—Ü–∏–∏
    _, minId = min((value, id) for (id, value) in enumerate(Z))
    X = X[:]
    Y = Y[:]
    Z = Z[:]

    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)

    return X, Y, Z

def evoSteps(X, Y, stepsNum = 4): # –®–∞–≥–∏ —ç–≤–æ–ª—é—Ü–∏–∏ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 4 —à–∞–≥–∞
    results = []

    for i in range(4):
        arrZ = [qZ(x, Y[i]) for i, x in enumerate(X)]

        X, Y, Z = evoStep(X, Y, arrZ)

        X, Y = exchangeSheme(X, Y, sorting(Z))

        results.append([X, Y, qSumZ(arrZ), arrZ])

    return X, Y, results

X = [-2, -1, 0, 1]
Y = [-2, -1, 0, 1]

results = evoSteps(X, Y)

for i in range(len(results[2])):
    print(f'max_{i + 1}_step: {results[2][i][2]}')

qualityArrZ = []
for i in range(len(results[2])):
    qualityArrZ += results[2][i][3]

print(f'max Z:       {max(qualityArrZ)}')
```

```python
# –ü—Ä–∏–º–µ—Ä 1.2.1

import networkx as nx
from math import e

distances = [(1, 2, 20),
            (1, 3, 40),
            (1, 4, 42),
            (1, 5, 33),
            (1, 6, 21),
            (2, 3, 26),
            (2, 4, 38),
            (2, 5, 42),
            (2, 6, 17),
            (3, 4, 22),
            (3, 5, 43),
            (3, 6, 21),
            (4, 5, 27),
            (4, 6, 22),
            (5, 6, 26)] # –î–ª–∏–Ω—ã —Ä—ë–±–µ—Ä

V = [1, 4, 5, 2, 6, 3, 1] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞
Z = [(3, 4),
     (4, 6),
     (5, 2),
     (6, 2)] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω –≤–µ—Ä—à–∏–Ω
P = [49, 54, 43, 54] # –°–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞, –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å–µ —Å—á—ë—Ç–∞

T = 100 # –ù–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞

# –§—É–Ω–∫—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
def probability(delta, T):
    return 100 * e ** (-delta / T)

# –§—É–Ω–∫—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã
def reductTemp(prevT):
    nextT = 0.5 * prevT

    return nextT

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞
def edgeLength(i, j, distances, roundTrip = True):
    if roundTrip:
        return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] == i and item[0] == j) else -1) for item in distances])
    else:
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –º–∞—Ä—à—Ä—É—Ç–∞
def routeLength(V, distances):
    edges = []

    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i], V[i + 1], distances))

    return sum(edges)

# –û–¥–Ω–æ–∫—Ä–∞—Ç–Ω–∞—è –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏
def routeOneReplacement(arrV, Z, replacementByName = True):
    decrement = 1 if replacementByName else 0

    arrV[Z[0] - decrement], arrV[Z[1] - decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]

    return arrV

# –ü–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏
def routeReplacement(V, Z):
    for z in Z:
        V = routeOneReplacement(V, z)
    return V

# –í—ã–±–æ—Ä –Ω—É–∂–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances) # –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –ø—É—Ç–∏
    arrSum = [sumLength] # –º–∞—Å—Å–∏–≤ —Å—É–º–º –¥–ª–∏–Ω
    #—Ü–∏–∫–ª –º–µ—Ç–æ–¥–æ–º –æ–±–∂–∏–≥–∞
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i]) #–Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        newS = routeLength(newV, distances) #–¥–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
        arrSum.append(newS)
        deltaS = newS - sumLength

        if deltaS > 0:
            p = probability(deltaS, T)

            if p > P[i]:
                V = newV
                sumLength = newS
            else:
                V = newV
                sumLength = newS
            T = reductTemp(T)

    return V, arrSum

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –º–∞—Ä—à—Ä—É—Ç—É

def drawRouteGraph(distances, bestRoute):
    newDistances = []
    for i in range(len(bestRoute) - 1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1] or distance[1] == bestRoute[i] and distance[0] == bestRoute[i + 1]:
                newDistances.append(distance)

    graph = nx.Graph() # –°–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞

    graph.add_weighted_edges_from(newDistances) # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä—ë–±–µ—Ä

    # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
    nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)

bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}')
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

drawRouteGraph(distances, bestRoute) # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
```

```python
# –ó–∞–¥–∞–Ω–∏–µ
# –ù–∞–π—Ç–∏ –¥–ª–∏–Ω—É –≥–∞–º–∏–ª—å—Ç–æ–Ω–æ–≤–∞ —Ü–∏–∫–ª–∞ –≤ –ø–æ–ª–Ω–æ–º –≥—Ä–∞—Ñ–µ –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä—ë—Ö —Ü–∏–∫–ª–æ–≤.
# –í–∞—Ä–∏–∞–Ω—Ç 8
import networkx as nx
from math import e

distances = [(1, 2, 26),
            (1, 3, 42),
            (1, 4, 44),
            (1, 5, 31),
            (1, 6, 24),
            (2, 3, 20),
            (2, 4, 34),
            (2, 5, 40),
            (2, 6, 15),
            (3, 4, 23),
            (3, 5, 43),
            (3, 6, 20),
            (4, 5, 27),
            (4, 6, 22),
            (5, 6, 26)] # –î–ª–∏–Ω—ã —Ä—ë–±–µ—Ä

V = [1, 2, 3, 4, 5, 6, 1] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞
Z = [(3, 4),
     (4, 6),
     (5, 6),
     (6, 2)] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω –≤–µ—Ä—à–∏–Ω
P = [90, 45, 43, 31] # –°–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞, –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å–µ —Å—á—ë—Ç–∞

T = 100 # –ù–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞

# –§—É–Ω–∫—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
def probability(delta, T):
    return 100 * e ** (-delta / T)

# –§—É–Ω–∫—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã
def reductTemp(prevT):
    nextT = 0.5 * prevT

    return nextT

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
# graph = nx.Graph() # –°–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞
# graph.add_weighted_edges_from(distances) # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä—ë–±–µ—Ä

# # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
# nx.draw_kamada_kawai(graph, node_color = '#fd7258', node_size = 2000, with_labels = True)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞
def edgeLength(i, j, distances, roundTrip = True):
    if roundTrip:
        return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] == i and item[0] == j) else -1) for item in distances])
    else:
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –º–∞—Ä—à—Ä—É—Ç–∞
def routeLength(V, distances):
    edges = []

    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i], V[i + 1], distances))

    return sum(edges)

# –û–¥–Ω–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏
def routeOneReplacement(arrV, Z, replacementByName = True):
    decrement = 1 if replacementByName else 0

    arrV[Z[0] - decrement], arrV[Z[1] - decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]

    return arrV

# –ü–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏
def routeReplacement(V, Z):
    for z in Z:
        V = routeOneReplacement(V, z)
    return V

# –í—ã–±–æ—Ä –Ω—É–∂–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances) # –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –ø—É—Ç–∏
    arrSum = [sumLength] # –º–∞—Å—Å–∏–≤ —Å—É–º–º –¥–ª–∏–Ω

    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i])
        newS = routeLength(newV, distances)
        arrSum.append(newS)
        deltaS = newS - sumLength

        if deltaS > 0:
            p = probability(deltaS, T)

            if p > P[i]:
                V = newV
                sumLength = newS
            else:
                V = newV
                sumLength = newS
            T = reductTemp(T)

    return V, arrSum

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –º–∞—Ä—à—Ä—É—Ç—É

def drawRouteGraph(distances, bestRoute):
    newDistances = []
    for i in range(len(bestRoute) - 1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1] or distance[1] == bestRoute[i] and distance[0] == bestRoute[i + 1]:
                newDistances.append(distance)

    graph = nx.Graph() # –°–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞

    graph.add_weighted_edges_from(newDistances) # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä—ë–±–µ—Ä

    # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏

    nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)

bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}')
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

drawRouteGraph(distances, bestRoute) # –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
```
